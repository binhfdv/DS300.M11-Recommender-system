```python
from pyvi import ViTokenizer, ViPosTagger
import string
# Define the function to remove the punctuation
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text

def handle_text(text):
  #remove punctuation:
  temp = remove_punctuations(text)

  #word tokenize:
  temp = ViTokenizer.tokenize(temp)

  #lowercase:
  temp = temp.lower()

  return temp

tokenized_text_feature = []
for sentence in text_feature:
  tokenized_text_feature.append(handle_text(sentence))
```

```python
from gensim.models import Word2Vec

sentences = [[item for item in doc.split()] for doc in tokenized_text_feature]
model = Word2Vec(sentences, min_count = 1, size = 150, window = 10, sg = 1, workers = 8)
model.train(sentences, total_examples = model.corpus_count, epochs = 10)
```

```python
import numpy as np

def get_artist_trackname_vector(user1_csv):

  user1_artist_trackname = []

  for i in range(len(user1_csv)):
    artist_vector = model.wv[[item for item in handle_text(user1_csv['artists'][i]).split()]]
    artist_vector = np.mean(artist_vector, axis=0)

    track_name_vector = model.wv[[item for item in handle_text(user1_csv['track_name'][i]).split()]]
    track_name_vector = np.mean(track_name_vector, axis=0)

    general = np.concatenate((artist_vector, track_name_vector), axis=0)
    user1_artist_trackname.append(general)
  return user1_artist_trackname

user1_train_artist_trackname = get_artist_trackname_vector(user1_train_csv)
user2_train_artist_trackname = get_artist_trackname_vector(user2_train_csv)
user3_train_artist_trackname = get_artist_trackname_vector(user3_train_csv)
```

## Light GBM:

```python
import time
import lightgbm as lgbm
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import log_loss, f1_score, accuracy_score


def train_LightGBM(X, y):

  N_SPLITS = 10
  strat_kf = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=42)

  LOSS = np.empty(N_SPLITS)
  f1_SCORE = np.empty(N_SPLITS)
  Accuracy = np.empty(N_SPLITS)

  for idx, (train_idx, test_idx) in enumerate(strat_kf.split(X, y)):
      print("=" * 12 + f"Training fold {idx+1}" + 12 * "=")
      start = time.time()

      X_train, X_val = X.iloc[train_idx], X.iloc[test_idx]
      y_train, y_val = y[train_idx], y[test_idx]
      eval_set = [(X_val, y_val)]

      lgbm_clf = lgbm.LGBMClassifier(n_estimators=10000)
      lgbm_clf.fit(
          X_train,
          y_train,
          eval_set=eval_set,
          early_stopping_rounds=200,
          eval_metric="binary_logloss",
          verbose=False,
      )

      pickle.dump(lgbm_clf, open('/content/drive/MyDrive/SEMESTER 7/RECOMMENDATION SYSTEM/Spotify_RecSys/Models/saved_models/LightGBM/lightgbm_model_%s.pickle' % str(idx+1), 'wb'))

      preds = lgbm_clf.predict_proba(X_val)
      loss = log_loss(y_val, preds)
      LOSS[idx] = loss

      print(np.argmax(preds, axis=-1))
      f1 = f1_score(y_val.tolist(), np.argmax(preds, axis=-1))
      f1_SCORE[idx] = f1

      acc = accuracy_score(y_val.tolist(), np.argmax(preds, axis=-1))
      Accuracy[idx] = acc
      runtime = time.time() - start
      print(f"Fold {idx+1} finished with loss: {loss:.5f} in {runtime:.2f} seconds.\n")
      print(f"Fold {idx+1} finished with f1 score: {f1:.5f} in {runtime:.2f} seconds.\n")
      print(f"Fold {idx+1} finished with accuracy: {acc:.5f} in {runtime:.2f} seconds.\n")

  print(f'Average f1 score of {N_SPLITS} folds: {np.mean(f1_SCORE, axis=0)}')
  print(f'Average accuracy of {N_SPLITS} folds: {np.mean(Accuracy, axis=0)}')
```

```python
# user 1
X = user1_X_train
y = user1_train_csv['label']
train_LightGBM(X, y)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import log_loss, f1_score, accuracy_score
import pickle
import time



def train_BoostingDT(X, y):
  N_SPLITS = 10
  strat_kf = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=42)

  LOSS = np.empty(N_SPLITS)
  f1_SCORE = np.empty(N_SPLITS)
  Accuracy = np.empty(N_SPLITS)

  for idx, (train_idx, test_idx) in enumerate(strat_kf.split(X, y)):
      print("=" * 12 + f"Training fold {idx+1}" + 12 * "=")
      start = time.time()

      X_train, X_val = X.iloc[train_idx], X.iloc[test_idx]
      y_train, y_val = y[train_idx], y[test_idx]
      eval_set = [(X_val, y_val)]

      # create model
      BoostingDT = AdaBoostClassifier(DecisionTreeClassifier(min_samples_split=10,max_depth=4),n_estimators=10,learning_rate=0.6)

      BoostingDT.fit(
          X_train,
          y_train)
      
      filename = '/content/drive/MyDrive/SEMESTER 7/RECOMMENDATION SYSTEM/Spotify_RecSys/Models/saved_models/BoostingDT/BoostingDecisionTree_model_%s.pkl' % str(idx+1)
      pickle.dump(BoostingDT, open(filename, 'wb'))

      preds = BoostingDT.predict_proba(X_val)
      loss = log_loss(y_val, preds)
      LOSS[idx] = loss

      f1 = f1_score(y_val.tolist(), np.argmax(preds, axis=-1))
      f1_SCORE[idx] = f1

      acc = accuracy_score(y_val.tolist(), np.argmax(preds, axis=-1))
      Accuracy[idx] = acc
      runtime = time.time() - start
      print(f"Fold {idx+1} finished with loss: {loss:.5f} in {runtime:.2f} seconds.\n")
      print(f"Fold {idx+1} finished with f1 score: {f1:.5f} in {runtime:.2f} seconds.\n")
      print(f"Fold {idx+1} finished with accuracy: {acc:.5f} in {runtime:.2f} seconds.\n")

  print(f'Average f1 score of {N_SPLITS} folds: {np.mean(f1_SCORE, axis=0)}')
  print(f'Average accuracy of {N_SPLITS} folds: {np.mean(Accuracy, axis=0)}')
```
